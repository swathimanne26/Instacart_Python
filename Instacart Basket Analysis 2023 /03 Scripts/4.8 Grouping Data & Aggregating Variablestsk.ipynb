#6 Create a spending flag for each user based on the average price across all their orders

```python
 #Creating 'average_price' column to show customers based on their mean prices across products
ords_prods_merge_df['average_price'] = ords_prods_merge_df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge_df.head()
```

```python
## creating "Spending_flag" for average price < 10 as low spender
ords_prods_merge_df.loc[ords_prods_merge_df['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#creating "Spending_flag" for average price >= 10 as High spender
ords_prods_merge_df.loc[ords_prods_merge_df['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge_df['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge_df.head()
```

```python
#7 #7 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
ords_prods_merge_df['median_order_frequency'] = ords_prods_merge_df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge_df.head()
```

```python
# Creating "order_frequency_flag" for median_orders > 20 as "Non-frequent customer"
ords_prods_merge_df.loc[ords_prods_merge_df['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating "order_frequency_flag" for median orders > 10 & <= 20 as "Regular customer"
ords_prods_merge_df.loc[(ords_prods_merge_df['median_order_frequency'] > 10) & (ords_prods_merge_df['median_order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer' 
```

```python
# Creating "order_frequency_flag" for median orders < 10 as "Frequent customer"
ords_prods_merge_df.loc[ords_prods_merge_df['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# checking "order_frequency_flag" frequency
ords_prods_merge_df['order_frequency_flag'].value_counts(dropna = False)
```
