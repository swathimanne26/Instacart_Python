# 4 Compare customers behaviour across geographic regions 

```python
##creating 'region' columns
ords_prods_cust_mergedspl_df['state'].value_counts(dropna = False)
```


##All states will be divided by regions according to Wikipedia list: https://simple.wikipedia.org/wiki/List_of_regions_of_the_United_States 

```python
# Create 'regions' list to assign a region for each state
regions = []

for value in ords_prods_cust_mergedspl_df['state']:
  if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
    regions.append('Northeast')
  elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
    regions.append('Midwest')
  elif value in ['Idaho','Montana','Wyoming','Nevada','Utah','Colorado','Arizona','New Mexico']:
    regions.append('West')
  else:
    regions.append('South')
```

```python
# Create 'regions' column within dataframe
ords_prods_cust_mergedspl_df['regions'] = regions
```

```python
# Check the output
ords_prods_cust_mergedspl_df.head()
```

```python
# Check frequency of values within 'regions' column
ords_prods_cust_mergedspl_df['regions'].value_counts()
```

```python
##Compare Customer spendings across regions
```

```python
 # Create crosstab between 'regions' and 'spending_flag' columns
crosstab_regn_spendg = pd.crosstab(ords_prods_cust_mergedspl_df['regions'],ords_prods_cust_mergedspl_df['spending_flag'], dropna = False)
```

```python
# Check the output
crosstab_regn_spendg
```

Northeast and West regions have approximately 16% of high spenders.
West has the lowest percentage of low spenders, which is 15,6%.
South region has the highest number of high- and low-spending customers, each group has 43% of total

# Create an exclusion flag for low - activity customers and exclude them from the data

```python
 # Set first condition (low-activity customers if orders number <5)
ords_prods_cust_mergedspl_df.loc[ords_prods_cust_mergedspl_df['max_order'] < 5, 'user_activity_flag'] = 'Low activity'
```

```python
 # Set second condition (regular customers if orders number >=5)
ords_prods_cust_mergedspl_df.loc[ords_prods_cust_mergedspl_df['max_order'] >= 5, 'user_activity_flag'] = 'Normal activity'
```

```python
# Check the output

ords_prods_cust_mergedspl_df['user_activity_flag'].value_counts(dropna = False)
```

```python
###Removing low-activity customers from dataframe
```

```python
# Create subset df with only low-activity customers
df_low_activity = ords_prods_cust_mergedspl_df[ords_prods_cust_mergedspl_df['user_activity_flag'] == 'Low activity']
```

```python
# Check the output
df_low_activity['user_activity_flag'].value_counts()
```

```python
# Create subset df with only normal-activity customers
df_normal_activity = ords_prods_cust_mergedspl_df[ords_prods_cust_mergedspl_df['user_activity_flag'] == 'Normal activity']
```

```python
# Check the output
df_normal_activity['user_activity_flag'].value_counts()
```
